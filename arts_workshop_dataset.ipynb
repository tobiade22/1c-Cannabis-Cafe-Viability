```python
df = pd.read_csv('courses.csv')
```

```python
# Replace 'wks' with empty string
df['Length'] = df['Length'].str.replace('wks', '')

# Replace '6 wks' with '6' and '1 day' with '1'
df['Length'] = df['Length'].str.replace('6 wks', '6').str.replace('1 day', '1').str.replace('2 days', '2')

# Convert the 'Length' column to integer
df['Length'] = df['Length'].astype(int)
```

```python
# Remove the '£' sign from the 'Price' column
df['Price'] = df['Price'].str.replace('£', '')

# Convert the 'Price' column to float (to handle decimal values) and then to integer
df['Price'] = df['Price'].astype(float).astype(int)
```

```python
df['Price per Day (£)'] = round(df['Price'] / df['Length'], 2)
```
